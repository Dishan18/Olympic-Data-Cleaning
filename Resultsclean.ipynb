# Results Data Cleaning

```python
import pandas as pd
results = pd.read_csv('./assets/results.csv')
```

```python
df = results.copy()
df.head()
```

```python
df[['Year', 'Season']] = df['Games'].str.extract(r'(\d{4}) (Summer|Winter)', expand=True)
df.head()
```

```python
df['new_medal'] = df["Medal"].map({"Bronze": 3, "Silver": 2, "Gold": 1})
df.head()
```

```python
df['position'] = df['Pos'].str.extract(r'(\d+)')
df['tied'] = df['Pos'].str.contains("=")
df['Event'] = df['Event'].str.slice(0, -9)
df.head()
```

```python
df.columns
```

```python
c_k = ['athlete_id', 'NOC','Event' ,'Discipline', 'Nationality', 'Year', 'Season', 'As', 'Team', 'Medal', 'position', 'tied']
df = df[c_k]
df.head()
```

```python
df.info()
```

```python
df['Year'] = pd.to_numeric(df['Year'])
df['position'] = pd.to_numeric(df['position'])
df['tied'] = df['tied'].astype(bool)
```

```python
df.info()
```

```python
df.head()
```

```python
df.to_csv('./results/results_cleaned.csv', index=False)
```
